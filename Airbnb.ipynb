#handling missing values



```python
df['last review'] = pd.to_datetime(df["last review"],errors = 'coerce')
```

```python
df["last review"]
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0 ,'last review':df['last review'].min()},inplace = True)
```

```python
df.dropna(subset = ['NAME','host name'],inplace=True)
```

```python
df[['NAME', 'host name']]
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns = ["license","house_rules"],errors = 'ignore')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#remove dollar sign in dollar and convert ot float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

# REMOVE DUPLICATES


```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

#Visualisation

# What is the ditribution of listing prices?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='red')

plt.title("Distribution of listing price")
plt.xlabel('price')
plt.ylabel("frequerncy")
plt.show()
```

# how a different room type distributed


```python
df['room type']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.countplot(x='room type', data=df, color='hotpink')
plt.title("room type distribution")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

sns.countplot(
    y='neighbourhood group',
    data=df,
    color='lightgreen',
    order=df['neighbourhood group'].value_counts().index
)

plt.title('Number of Listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')

plt.show()
```

# WHAT IS THE RELATIONSHIP BETWEEN PRICE AND ROOM TYPE


```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type' ,y='price',hue='room type',data =df,palette='Set1')
plt.title("relationship between price and room type")
plt.xlabel("room type")
plt.ylabel("price")
plt.show()
```
